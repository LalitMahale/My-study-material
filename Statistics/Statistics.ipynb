### 2. What is the difference between inferential and descriptive statistics?

- **Inferential Statistics :** 
  
  - In inferential statistics we make prediction based population based on sample data.
  - In inferetial statistics we do mesure of t-test, hypothesis testing, regression **that allow conclusions about a population based on sample data.**

- **Descriptive Statistics :**
    - In descriptive statistics we **Sumerize and describe the dataset based on characteristics of dataset**
    - in Descriptive statistics we focus only on sample data.
    - **measures like mean, mode , median, standard deviation, graphical representation which sumerize the data**

###### In short, descriptive statistics describe "what" the data shows, while inferential statistics help determine the "why" or make predictions about a larger group.

## 5. What are the types of sampling.

In simple random sampling is select randomly from population. and there is fix probablity to be selected.


#### 2. Clustering Sampling:
In this method we divide entire population into clusters. then the cluster are identified in parameters eg sex, gender, age , location.


#### 3. Stratified Sampling :
In stratified sampling we divide entire population in different or unique groups that represet the entire population.
while sampling these group can be orgnized and then dawn a sample from each group seperatly.

#### 4. Systematic Sampling 
In this method we systematically choose the sample based on fixed, period or interval called sampling interval.
The sampling interval is calculated by divide the population by desired sample size. This type of sampling method has a predefine range. 
since it has predefine range it is less time consuming.
