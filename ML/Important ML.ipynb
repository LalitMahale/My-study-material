## What is the trade-off between bias and variance?
**Bias** and **variance** are two important concepts in machine learning that help us understand the sources of error in predictive models:

### Bias
- **Definition**: Bias refers to the error introduced by approximating a real-world problem, which may be complex, with a simplified model.
- **Characteristics**: 
  - High bias can lead to underfitting, where the model is too simplistic to capture the underlying patterns in the data.
  - Models with high bias often make strong assumptions and are less flexible.
- **Example**: A linear model trying to fit a highly nonlinear relationship may show high bias.

### Variance
- **Definition**: Variance refers to the model's sensitivity to fluctuations in the training data. It measures how much the model's predictions would change if it were trained on a different dataset.
- **Characteristics**: 
  - High variance can lead to overfitting, where the model captures noise in the training data rather than the true underlying patterns.
  - Models with high variance are often very complex and flexible.
- **Example**: A decision tree that perfectly fits the training data, capturing every detail, may have high variance.

### Bias-Variance Tradeoff
- The goal in machine learning is to find a balance between bias and variance to minimize the total error:
  - **Low Bias, High Variance**: The model is complex and fits the training data well but may perform poorly on unseen data.
  - **High Bias, Low Variance**: The model is too simple and may not perform well on either training or test data.
  
Finding the right model complexity is crucial to achieving a good generalization on new data.

## What are the common ways to handle missing data in a dataset?
- First **Delete the rows which contain missing Values**
- If it is not possible to remove than fill with **Mean / Median / Mode**
- Use algorith which support missing values eg . **Decision Tree or Random Forest**

## What is Bagging and Boosting?
- Bagging is a ensemble learning which is used for improving unstable estimation or classification scheme
- Boosting used to reduce the bias of the combined models.

## Regularization 

Regularization is used to address **overfitting problems** as it penalizes the loss function by adding a multiple of an L1 (LASSO) or an L2 (Ridge) norm of weights vector w.
